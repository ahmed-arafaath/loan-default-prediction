```python
#Importing all the essential packages
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
df = pd.read_csv('LoanDefaultData.csv')
df.info()
```

```python
#Selecting columns for training the model
x = df[['AP001', 'AP003','AP004','TD001','TD002','TD005','TD006','TD009','TD010','TD013','TD014','TD015', 'CR015', 'CR017', 'CR019']]
```

```python
#Renaming the column for readability
x.rename(columns={'AP001': 'Age', 'AP003': 'Education', 'AP004': 'Loan Term', 'TD001': 'LAST_7Day_P2P', 
                  'TD002': 'LAST_7Day_SMALL_LOAN', 'TD005': 'LAST_1MON_P2P', 'TD006': 'LAST_1MON_SMALL_LOAN', 
                  'TD009': 'LAST_3MON_P2P', 'TD010': 'LAST_3MON_SMALL_LOAN', 'TD013': 'LAST_6MON_P2P',
                 'TD014': 'LAST_6MON_SMALL_LOAN', 'TD015': 'LAST_6MON_BANK', 'CR015': 'MONTH_CREDIT_CARD_MOB_MAX',
                  'CR017': 'SCORE_DEBIT_CARD_TOTAL_AMT','CR019': 'SCORE_SINGLE_DEBIT_CARD_LIMIT'}, inplace=True)
```

I selected numerous columns to train my model for Loan Default predictor, and the explanation as follows,
1. Age - I selected age to determine the individual's experience level and their capacity of income.
2. Education - I selected education to determine the individual's expected income based on education.
3. Loan term - I selected loan term to take in to consideration the monthly repayment based on their income and duration.
4. Last_7day_P2P, Last_1Mon, Last_3Mon, Last_6Mon - I selected this column to take into consideration how often the individual lends money.
5. Last_7Dat_Small_Loan, Last_1Mon, Last_3Mon, Last_6Mon - to take into consideration how often the individual seeks loan
6. Last_6Mon_Bank - to check what is average bank balance
7. Month_Credit_Card_Max - How often individual reaches their credit card max.
8. Score_Single_Debit_Card_Limit - Take into consideration of the individual's debit card overdraft

```python
#dividing the dataset into train and test for model implementation.
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2)
X_train.head()
```

```python
Y_train.head()
```

## KNN Algorithm 

```python
#Implementing KNN Classifies
from sklearn import neighbors, linear_model, metrics, naive_bayes
knn = neighbors.KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train, Y_train)
```

```python
y_pred = knn.predict(X_test)
y_pred
```

```python
print(metrics.classification_report(Y_test,y_pred))
```

```python
metrics.accuracy_score(Y_test, y_pred)
```

## Logistic Regression

```python
#implementing logistic regression
lr_model = linear_model.LogisticRegression()
lr_model.fit(X_train, Y_train)
```

```python
predicted_values = lr_model.predict(X_test)
lr_accuracy = metrics.accuracy_score(predicted_values, Y_test)
print(lr_accuracy)
```

```python
print(metrics.classification_report(Y_test,predicted_values))
```

```python
predicted_values
```

## Naive Bayes Classifier

```python
#Implementing Naive Bayes Classifier
nb_model = naive_bayes.GaussianNB()
nb_model.fit(X_train, Y_train)
```

```python
nb_y_model = nb_model.predict(X_test)
nb_accuracy = metrics.accuracy_score(Y_test, nb_y_model)
print(nb_accuracy)
```

```python
nb_y_model
```

```python
print(metrics.classification_report(Y_test,nb_y_model))
```

# From the above 3 models, it seems like logistic regression has the highest accuracy and the model seem to be accurate at predicting non-loan defaulter but not accurate enough to predict loan defaulters.

## Approach 2 - Balancing the data

```python
gk = df.loc[df['loan_default'] == 0] #rows with borrowers who did not default loan
gk
```

```python
gk = gk.loc[:20000] #extracting only 20,000 rows
gk.info()
```

```python
gk1 = df.loc[df['loan_default'] == 1] #rows of borrowers who defaulted loan
gk1.info()
```

```python
#Merging the datasets
frame = [gk, gk1]
alt_df = pd.concat(frame)
alt_df
```

```python
alt_df.info() #reduced dataset with balance
```

```python
alt_df = alt_df.sample(frac=1).reset_index(drop=True) #random shuffling of the dataset to reduce bias in test and train
alt_df
```

```python
#selecting dataset for model
y = alt_df.loan_default
x = alt_df[['AP004','TD001','TD002','TD005','TD006','TD009','TD010','TD013','TD014','TD015', 'CR015', 'CR017', 'CR019']]
```

```python
x.head()
```

```python
X2_train, X2_test, Y2_train, Y2_test = train_test_split(x, y, test_size=0.2)
X2_train.head()
```
